math

Se utiliza para realizar operaciones matemáticas básicas con números escalares.

```python
import math

x=16
#calcular la raiz cuadrada de x
resultado = math.sqrt (x)  #funcion raiz cuadrada
print("la raiz cuadrada de 16 es:", resultado)
```

```python
import numpy as np
m= np.array([1, 2, 3, 4])
# Realizar una operacion sobre el array
squared = np.sqrt(m) #Raiz cuadrada de cada elemento
print("Las raices cuadradas de los elementos son:", squared)
```

matplotlib.pyplot
Se utiliza para crear gráficos y visualizar datos.


```python
import matplotlib.pyplot as plt
x=np.linspace (-10, 10, 100) #Generar 100 puntos de -10 a 10
y= x**2 #Funcion cuadratica
plt.plot(x, y)
plt.title("Grafico de f(x) x^2")
plt.xlabel("x")
plt.xlabel("f(x)")
plt.grid(True)
plt.show()
```
